```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
eco = pd.read_excel('/content/converted_file.xlsx')
```

```python
eco.head()
```

```python
eco.isnull().sum()
```

```python
eco.dropna(inplace = True)
```

```python
eco.isnull().sum()
```

```python
eco.info()
```

```python
lr = LogisticRegression()
```

```python
ind = eco[['Customer ID','Age','Items Purchased']]
dep = eco['Satisfaction Level']
```

```python
lr.fit(ind,dep)
```

```python
print("your predicted discount is",lr.predict([[3,34,7]]))
```

```python
lr.score(ind,dep)
```

```python
id = int(input("enter the customer id:"))
age = int(input("enter your age:"))
item = int(input("enter number of item you purchased:"))
pval = lr.predict([[id,age,item]])
print(pval)
```
